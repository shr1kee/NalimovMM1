```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img =cv2.imread("C:/Users/Acer/opencvProj/image.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
img_gr = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
print(type(img_gr))
plt.imshow(img_gr, cmap="gray")
plt.show()
newim = img_gr.copy()
k = int(input())
if(k%2==0):
    k=k+1
intK=k//2
for i in range(intK,img_gr.shape[0]-intK):
    for j in range(intK,img_gr.shape[1]-intK):
        p=[]
        for x in range(i-intK,i+intK):
            for y in range(j-intK,j+intK):
                p.append(img_gr[x,y])
        p.sort()
        newim[i,j]=p[len(p)//2+1]
plt.imshow(newim,cmap="gray")
plt.show()
        

        
```
